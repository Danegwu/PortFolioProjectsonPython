### GROUPING INTO MALE SENIOR CITIZEN AND FEMALE SENIOR CITIZEN

### 3. More Analysis Of the Data

#### a. Payment Methods Between Senior and Junior Citizen Subscribers

```python
seniors=df[df['SeniorCitizen']==1]
juniors=df[df['SeniorCitizen']==0]
fig, axes = plt.subplots(1, 2, figsize=(20, 7))
fig.suptitle('Variation of Payment Methods Between Senior and Junior Citizen Subscribers',fontsize=20)
sns.countplot(ax=axes[0], x='PaymentMethod', data=seniors)
axes[0].set_title("Senior Citizens",fontsize=15)
sns.countplot(ax=axes[1], x='PaymentMethod', data=juniors)
axes[1].set_title("Junior Citizens",fontsize=15)
```

Across both the junior and senior subscribers, electroic checks is the most popular payment method. Among the senior citizen subscribers, mailed check is the least used mode of payment while bank transfer and credit card are least used among the junior subscribers

#### b. Length of Contract between Senior and Junior Citizen 

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))
fig.suptitle('Comparison of Contract Lengths Between Senior and Junior Citizen Subscribers',fontsize=20)
sns.countplot(ax=axes[0], x='Contract', data=seniors)
axes[0].set_title("Senior Citizens",fontsize=15)
sns.countplot(ax=axes[1], x='Contract', data=juniors)
axes[1].set_title("Junior Citizens",fontsize=15)
```

Across both senior and non-senior citizens, month-to-month contract is mostly used contract. Two year and one year contracts are least used.

#### c. Movie Streams Between Senior and Junior Citizen

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))
fig.suptitle('Comparison of Movie Streaming Service Between Senior and Junior Citizen Subscribers',fontsize=20)
sns.countplot(ax=axes[0], x='StreamingMovies', data=seniors)
axes[0].set_title("Senior Citizens",fontsize=15)
sns.countplot(ax=axes[1], x='StreamingMovies', data=juniors)
axes[1].set_title("Junior Citizens",fontsize=15)
```

Most of the senior subscribers stream movies while the highest number of junior citizens do not use streaming services and there is also a high number of Junior Citizens who use streaming services.

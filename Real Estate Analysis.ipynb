```python
# Reducing the columns to the ones needed to implement the solution.
Sample1_df = df[['SaleDate','SalePrice']]  
Sample2_df = df[['StreetNameAndWay','SalePrice']]  
```

## Implementing the solution

### I. Total Sale per Quarter of the Real Estate

```python
# getting the first and last date in our dataframe
print(df['SaleDate'].min())
print(df['SaleDate'].max())
```

```python
# Getting the quarterly periods of the dates given 
Sample1_df['Quarter'] = Sample1_df['SaleDate'].dt.quarter
Sample1_df.head()
```

```python
# getting the required columns to a dataframe so as to plot
Sample1 = Sample1_df.groupby(['Quarter']).sum()
Sample1 = Sample1.reset_index()
Sample1
```

```python
# bar graph of the total sale prices according to the quartely periods. 
sns.set(style="darkgrid")
plt.figure(figsize=(10,10))
sns.barplot(x='Quarter',y='SalePrice',data=Sample1)
plt.title("Total Price Considering the Quarterly Periods",fontsize=18)
plt.xlabel("Quarterly Periods",fontsize=14)
plt.ylabel("Sale Price",fontsize=14)
plt.show()
```

It is evident that the fourth quarter is more productive in terms of sales totaling to 1887341730 dollars. It may be that the festivities bring about the high real estate sales in the 4th quarter. the 1st quarter follows closely then the third quarter and lastly the 2nd quarter.

### II. Average Sale price per Area of the Real Estate

```python
# getting the required columns to a dataframe so as to plot

Sample2 = Sample2_df.groupby(['StreetNameAndWay']).mean().round()
Sample2 = Sample2.reset_index()

# Sorting in descending order 
Sample2= Sample2.sort_values(['SalePrice'],ascending = False)

# converting the datatype of the sales column to int
Sample2['SalePrice'] = Sample2['SalePrice'].astype(int)
Sample2
```

```python
# plotting the average sale price per area of the real estate for the first 40 ranked areas according to mean sale price
sns.set(style="whitegrid")
plt.figure(figsize=(20,20))
sns.barplot(x='SalePrice',y='StreetNameAndWay',data=Sample2.head(40))
plt.title("Average Sale price per Area of the Real Estate",fontsize=20)
plt.xlabel("Sale Price",fontsize=15)
plt.ylabel("Real Estate Area",fontsize=15)
plt.show()
```

It is evident that Columbus Blvd has averagely the highest total sales of real estates with an average of 16800000 dollars for the sales. The first 40 locations represent areas with average sales of 967000 dollars and above.

### III. Which month has the biggest Sale?

```python
# creating a column for months so as to plot with it
Sample1_df['Month'] = pd.DatetimeIndex(Sample1_df['SaleDate']).month
Sample1_df['Year'] = pd.DatetimeIndex(Sample1_df['SaleDate']).year

Sample1_df
```

```python
Sample3 = Sample1_df.groupby(['Month'])['SalePrice'].sum()
Sample3
```

```python
# barplot for total sales per month of the year
plt.figure(figsize=(20,10))
sns.barplot(x ="Month", y ="SalePrice", hue ="Year", data=Sample1_df)
plt.title("Total Sale Price per month ",fontsize=18)
plt.xlabel("Months",fontsize=14)
plt.ylabel("Sale Price",fontsize=14)
plt.show()
```

December has a very high affinity for Real Estate sales with cumulative sales totalling to 1681501974 dollars. It is followed closely by January with a cumulative total sales of 588351895 dollars. July is third with a cumulative sales totalling to 329201813 dollars. the rest of the months are below the 300000000 dollar mark.

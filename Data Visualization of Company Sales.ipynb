```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales= pd.read_csv("company_sales_data.csv", encoding = 'unicode_escape')
sales
```

```python
sales.index
```

## Implementing the solution

## QUESTION 1

1.	Read Total profit of all months and show it using a line plot

```python
plt.plot(sales['month_number'], sales['total_profit'], color='blue')
plt.title('Company Profit per Month', fontsize=14)
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Total Profit', fontsize=14)
plt.show()
```

## QUESTION 2

2.	Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties: –
•	Line Style dotted and Line-color should be red
•	Show legend at the lower right location.
•	X label name = Month Number
•	Y label name = Sold units number
•	Add a circle marker.
•	Line marker color as read
•	Line width should be 3


```python
plt.plot(sales['month_number'], sales['total_units'], color='red', marker='o',markerfacecolor='k',linestyle='--',linewidth=3)
plt.title('Company Sales Data of Last Year', fontsize=14)
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sold Unit Number', fontsize=14)
plt.legend(['Profit Data of Last Year'],loc=4)
plt.show()
```

## QUESTION 3

3.	Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

```python
plt.plot(sales['facecream'], label = 'Face Cream Sales Data', marker='o',linewidth=3)
plt.plot(sales['facewash'], label = 'Face Wash Sales Data', marker='o',linewidth=3)
plt.plot(sales['toothpaste'], label = 'ToothPaste Sales Data',marker='o',linewidth=3)
plt.plot(sales['bathingsoap'], label = 'Bathing Soap Sales Data',marker='o',linewidth=3)
plt.plot(sales['shampoo'], label = 'Shampoo Sales Data', marker='o',linewidth=3)
plt.plot(sales['moisturizer'], label = 'Moisturizer Sales Data', marker='o',linewidth=3)
plt.title('Sales Data', fontsize=14)
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sales Units in Numbers', fontsize=14)
plt.legend(loc=2)
#plt.grid(True)
plt.show()
```

## QUESTION 4

4.	Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot. gridline style should “–“.


```python
plt.scatter(sales['month_number'], sales['toothpaste'], label='Tooth Paste Sales Data')
plt.title('Sales Data', fontsize=14)
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sales Units in Numbers', fontsize=14)
plt.legend(loc=2)
plt.grid(True, linewidth=1,linestyle="--")
plt.show()
```

## QUESTION 5

5.	Read face cream and facewash product sales data and show it using the bar chart.

	The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


```python
ax = sales.plot.bar(x='month_number',y=['facecream','facewash'],rot=0,figsize=(12,5),fontsize=10)
plt.ylabel('Sales Unit in Numbers')
plt.title('FaceWash and FaceCream Sales Data')
plt.legend(loc=2)
plt.grid(True, linewidth=1,linestyle="--")
```

## QUESTION 6

6.	Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
ax = sales.plot.bar(x='month_number',y=['bathingsoap'],rot=0,figsize=(12,5),fontsize=10)
plt.ylabel('Sales Unit in Numbers')
plt.title('Bathing Soap Sales Data')
# plt.legend(loc=2)
plt.grid(True, linewidth=1,linestyle="--")
```

## QUESTION 7

7.	Read the total profit of each month and show it using the histogram to see most common profit ranges

```python
plt.hist(sales['total_profit'])
plt.title('Company Sales Data of Last Year', fontsize=14)
profit = [150000,175000,200000,225000,250000,300000,350000,400000]
plt.xlabel('Profit Range in Dollar', fontsize=14)
plt.ylabel('Actual Profit in Dollar', fontsize=14)
plt.legend(['Profit Data'],loc=2)
plt.xticks(profit)
plt.show()
```

## QUESTION 8

8.	Calculate total sale data for last year for each product and show it using a Pie chart

Note: In Pie chart display Number of units sold per year for each product in percentage.


```python
data = ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesum   = [sales ['facecream'].sum(), sales ['facewash'].sum(), sales ['toothpaste'].sum(), 
         sales ['bathingsoap'].sum(), sales ['shampoo'].sum(), sales ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesum, data=data, autopct='%1.1f%%')
plt.legend(data,loc=4)
plt.title('Sales data')
plt.show()
```

## QUESTION 9

9.	Read Bathing soap facewash of all months and display it using the Subplot

```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(sales['month_number'],sales['bathingsoap'], label='Bathing Soap Sales Data',
              color='k', marker='o',linewidth=3)
axarr[0].set_title('Sales Data of BathingSoap')
axarr[1].plot(sales['month_number'],sales['facewash'], label='Face Wash Sales Data',color='r',
              marker='o',linewidth=3)
axarr[1].set_title('Sales Data of FaceWash')
plt.xticks(sales['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
```

## QUESTION 10

10.	Read all product sales data and show it using the stack plot

```python
plt.plot([],[],color='m', label='Face Cream', linewidth=3)
plt.plot([],[],color='c', label='Face wash', linewidth=3)
plt.plot([],[],color='r', label='Tooth paste', linewidth=3)
plt.plot([],[],color='k', label='Bathing soap', linewidth=3)
plt.plot([],[],color='g', label='Shampoo', linewidth=3)
plt.plot([],[],color='y', label='Moisturizer', linewidth=3)

plt.stackplot(sales['month_number'],sales['facecream'],sales['facewash'],sales['toothpaste'],
              sales['bathingsoap'],sales['shampoo'],sales['moisturizer'],
              colors=['m','c','r','k','g','y'])
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('All product sales data using stack plot')
plt.legend(loc=2)
plt.show()
```

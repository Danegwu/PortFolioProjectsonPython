### 12. Find the total number of Person having age between 17 to 48(inclusive) using between method

```python
# without using between method
print('The Number of Persons between the age of 17 and 48 is',sum((data['age']>=17) & (data['age']<=48)))
```

```python
# using between method
print('The Number of Persons between the age of 17 and 48 is',sum(data['age'].between(17,48)))
```
